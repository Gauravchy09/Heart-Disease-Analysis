23. Correlation heatmap: Use Matplotlibâ€™s imshow (or matshow) to display a heatmap of the correlation matrix for all numeric features. Include a color bar and appropriate axis labels for feature names.

A heatmap of a correlation matrix visually shows how strongly two variables are related to each other.

Darker/redder colors â†’ strong positive correlation (near +1).

Darker/bluer colors â†’ strong negative correlation (near -1).

Pale colors â†’ weak or no correlation (near 0).

```python
# Step 1: Calculate correlation matrix
corr_matrix = df.corr()

# Step 2: Create the heatmap
plt.figure(figsize=(10,8))
plt.imshow(corr_matrix, cmap='coolwarm', interpolation='nearest')  # 'coolwarm' color scale
plt.colorbar()  # color bar to show correlation values

# Step 3: Set axis ticks and labels
plt.xticks(ticks=np.arange(len(corr_matrix.columns)), labels=corr_matrix.columns, rotation=90)
plt.yticks(ticks=np.arange(len(corr_matrix.columns)), labels=corr_matrix.columns)

plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```

24. Scatter plot of a strong relationship: Identify a pair of non-target numeric features that appear strongly related (for example, from the correlation matrix). Create a scatter plot of these two features and optionally overlay a regression line computed with NumPy.

```python
# Scatter plot
x = df['age']
y = df['thalach']

plt.scatter(x, y, color='blue', label='Data points')

# Regression line (using NumPy)
m, b = np.polyfit(x, y, 1)  # Fit a 1st degree polynomial (line)
plt.plot(x, m*x + b, color='red', label='Best fit line')

plt.xlabel('Age')
plt.ylabel('Maximum Heart Rate (thalach)')
plt.title('Age vs Maximum Heart Rate')
plt.legend()
plt.grid()
plt.show()
```

25. Age vs. Blood Pressure by target: Make a scatter plot of age (x-axis) vs. resting blood pressure (trestbps, y-axis), using different colors or markers for patients with heart disease vs. those without. Do the two groups appear to differ?

```python
# Assuming 'target' column indicates if the patient has heart disease (1 for yes, 0 for no)
plt.figure(figsize=(10, 6))

# Scatter plot for patients with heart disease (target == 1) in red
plt.scatter(df[df['target'] == 1]['age'], df[df['target'] == 1]['trestbps'], 
            color='red', label='Heart Disease', alpha=0.6)

# Scatter plot for patients without heart disease (target == 0) in blue
plt.scatter(df[df['target'] == 0]['age'], df[df['target'] == 0]['trestbps'], 
            color='blue', label='No Heart Disease', alpha=0.6)

# Labels and title
plt.xlabel('Age')
plt.ylabel('Resting Blood Pressure (trestbps)')
plt.title('Age vs. Resting Blood Pressure by Heart Disease Target')

# Add legend
plt.legend()

# Show plot
plt.show()
```
